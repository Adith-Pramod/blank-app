```python
import os
import pandas as pd
import json

# This is the path returned from kagglehub
dataset_path = "/root/.cache/kagglehub/datasets/nikitamanaenkov/annotated-x-ray-angiography-dataset/versions/2"
image_dir = os.path.join(dataset_path, "data", "stenosis", "train", "images")
mask_dir = os.path.join(dataset_path, "data", "stenosis", "train", "masks")
label_file = os.path.join(dataset_path, "data", "stenosis", "train", "annotations", "train.json")

# Load the JSON file
with open(label_file, 'r') as f:
    data = json.load(f)

# Load the 'annotations' data into a DataFrame
df = pd.DataFrame(data['annotations'])

# Display the first few rows and the columns of the DataFrame
print(df.head())
print(df.columns)
```

```python
import cv2
import numpy as np
from sklearn.model_selection import train_test_split
import os
import pandas as pd
import json

def load_data_with_labels(annotations_df, images_data, image_dir, img_size=(256, 256)): # Removed mask_dir
    X_img, y_stenosis = [], [] # Removed X_mask and y_mask

    # Create a dictionary for quick lookup of image filenames by image_id
    image_files = {img['id']: img['file_name'] for img in images_data}

    print(f"Total annotations to process: {len(annotations_df)}")
    processed_count = 0

    # Category ID for 'stenosis' based on the categories data
    STENOSIS_CATEGORY_ID = 26

    for index, row in annotations_df.iterrows():
        image_id = row['image_id']
        fname = image_files.get(image_id)
        category_id = row['category_id']

        # Check if the annotation is for stenosis
        is_stenosis = (category_id == STENOSIS_CATEGORY_ID)

        if fname is None:
            print(f"Warning: Image file not found for image_id: {image_id}")
            continue

        img_path = os.path.join(image_dir, fname)

        if not os.path.exists(img_path):
            print(f"Warning: Image file not found at path: {img_path}")
            continue

        img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

        if img is None:
            print(f"Warning: Could not read image file at path: {img_path}")
            continue

        # Resize and normalize images
        img = cv2.resize(img, img_size) / 255.0

        X_img.append(img)
        # Append 1 if it's a stenosis annotation, 0 otherwise
        y_stenosis.append(1 if is_stenosis else 0)

        processed_count += 1

    print(f"Successfully processed {processed_count} annotations.")
    return (
        np.expand_dims(np.array(X_img), -1),
        np.array(y_stenosis)
    )


# Load the JSON file again to access the 'images' data
dataset_path = "/root/.cache/kagglehub/datasets/nikitamanaenkov/annotated-x-ray-angiography-dataset/versions/2"
label_file = os.path.join(dataset_path, "data", "stenosis", "train", "annotations", "train.json")
with open(label_file, 'r') as f:
    data = json.load(f)

# Get the images data
images_data = data['images']

# Use the already loaded annotations DataFrame 'df'
# Removed mask_dir argument in the function call
X, y_stenosis = load_data_with_labels(df, images_data, image_dir)

# Train-test split
# Adjusted train_test_split to only use X and y_stenosis
X_train, X_val, y_stenosis_train, y_stenosis_val = train_test_split(
    X, y_stenosis, test_size=0.2, random_state=42
)

print("Shape of X_train:", X_train.shape)
# Removed print for y_mask_train shape
print("Shape of y_stenosis_train:", y_stenosis_train.shape)
```

```python
import matplotlib.pyplot as plt
import numpy as np # Import numpy

def show_predictions(model, X, y_stenosis, class_names=["None", "Stenosis"]): # Removed y_mask and added default class_names
    preds_class = model.predict(X[:5]) # Removed mask prediction

    for i in range(5):
        plt.figure(figsize=(5, 5)) # Adjusted figure size
        plt.imshow(X[i].squeeze(), cmap='gray')
        plt.title(f"True Class: {class_names[y_stenosis[i]]}\nPredicted Class: {class_names[np.argmax(preds_class[i])]}") # Displaying true and predicted class
        plt.axis('off') # Hide axes
        plt.show()

# Adjusted the function call to match the modified function signature
show_predictions(model, X_val, y_stenosis_val)
```

```python
import tensorflow as tf
from tensorflow.keras import layers, Model

def build_model(input_shape=(256, 256, 1), num_classes=3):
    inputs = layers.Input(shape=input_shape)

    # Encoder
    c1 = layers.Conv2D(16, 3, activation='relu', padding='same')(inputs)
    p1 = layers.MaxPooling2D()(c1)
    c2 = layers.Conv2D(32, 3, activation='relu', padding='same')(p1)
    p2 = layers.MaxPooling2D()(c2)
    c3 = layers.Conv2D(64, 3, activation='relu', padding='same')(p2)

    # Classification head
    x = layers.GlobalAveragePooling2D()(c3)
    x = layers.Dense(64, activation='relu')(x)
    stenosis_output = layers.Dense(num_classes, activation='softmax', name="stenosis_output")(x)

    # Modified model with only classification output
    model = Model(inputs=inputs, outputs=stenosis_output)
    model.compile(
        optimizer='adam',
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )
    return model

model = build_model()
```
